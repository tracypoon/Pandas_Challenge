## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# List all the columns in the table
purchase_data.columns
```

```python
#retrieve by Item ID, Item Name and Price
item_df = purchase_data[["Item ID", "Item Name" , 'Price']]
item_df.head()
```

```python
#groupby Item ID and Item Name
Item_Group = item_df.groupby(['Item ID', 'Item Name'])
Item_Group["Item Name"].count()
Item_Analysis = pd.DataFrame({"Purchase Count" : Item_Group["Item Name"].count(),
                              "Item Price" : Item_Group["Price"].mean(),
                              "Total Purchase Value": Item_Group["Price"].sum()})

#print the dataframe
Item_Analysis.head()
```

```python
#sort Purchase Count by descending order
sorted_df2 = Item_Analysis.sort_values(by="Purchase Count", ascending=False)


#display dataframe with dollar format
sorted_df2["Item Price"] = sorted_df2["Item Price"].map("${:,.2f}".format)
sorted_df2["Total Purchase Value"] = sorted_df2["Total Purchase Value"].map("${:,.2f}".format)
sorted_df2.head()
```
